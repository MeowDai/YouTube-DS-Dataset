# No CoT, No Exclusion Prompt

```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, ConfusionMatrixDisplay

# Database path
db_path = r'F:\classification\youtube_video_data.db'

# Connect to SQLite database
conn = sqlite3.connect(db_path)

# Read gpt4o_results table
df = pd.read_sql_query("SELECT video_id, true_label, gpt_prediction FROM gpt4o_results WHERE video_id IN (SELECT video_id FROM manual_labels)", conn)
conn.close()

# Calculate classification evaluation metrics
accuracy = accuracy_score(df['true_label'], df['gpt_prediction'])
precision = precision_score(df['true_label'], df['gpt_prediction'], average='binary') 
recall = recall_score(df['true_label'], df['gpt_prediction'], average='binary')
f1 = f1_score(df['true_label'], df['gpt_prediction'], average='binary')

# Display evaluation metrics
print(f'Accuracy: {accuracy:.4f}')
print(f'Precision: {precision:.4f}')
print(f'Recall: {recall:.4f}')
print(f'F1 Score: {f1:.4f}')

# Calculate and plot confusion matrix
cm = confusion_matrix(df['true_label'], df['gpt_prediction'])
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot(cmap=plt.cm.Blues)
plt.title('Confusion Matrix')
plt.show()

# Count prediction correctness
df['correct'] = df['true_label'] == df['gpt_prediction']
correct_counts = df['correct'].value_counts()
plt.bar(['Correct', 'Incorrect'], correct_counts, color=['green', 'red'])
plt.ylabel('Count')
plt.title('Correct vs Incorrect Predictions')
plt.show()
```
